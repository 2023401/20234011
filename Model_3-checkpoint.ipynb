## Unsupervised ML Model

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Selecting relevant features for clustering
X_unsupervised = df[['cong', 'dsat', 'cong_pc', 'dsat_pc']]

# Standardize features
scaler_unsupervised = StandardScaler()
X_unsupervised_scaled = scaler_unsupervised.fit_transform(X_unsupervised)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)  # we can choose the number of clusters based on our data
df['cluster'] = kmeans.fit_predict(X_unsupervised_scaled)

# Visualizing the clusters
plt.figure(figsize=(12, 6))
sns.scatterplot(x='cong', y='dsat', hue='cluster', data=df, palette='viridis', s=50)
plt.title('K-Means Clustering of Congestion vs DSAT')
plt.xlabel('Congestion')
plt.ylabel('DSAT')
plt.show()
```
