```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Traffic_Flow.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# Summary statistics for numerical columns
print(df.describe())
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Visualize distribution of numerical features
plt.figure(figsize=(12, 8))
sns.histplot(df['flow'], bins=30, kde=True)
plt.title('Distribution of Flow')
plt.show()
```

```python
# Visualize correlations between numerical features
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Boxplot for flow by site
plt.figure(figsize=(16, 8))
sns.boxplot(x='site', y='flow', data=df)
plt.title('Flow Distribution by Site')
plt.xticks(rotation=45)
plt.show()
```

```python
# Countplot for 'site'
plt.figure(figsize=(12, 6))
sns.countplot(x='site', data=df)
plt.title('Count of Observations by Site')
plt.xlabel('Site')
plt.ylabel('Count')
plt.show()
```

```python
# Pairplot for numerical features
sns.pairplot(df[['flow', 'cong', 'dsat']])
plt.suptitle('Pairplot of Numerical Features', y=1.02)
plt.show()
```

```python
# Violin plot for congestion percentage by site
plt.figure(figsize=(14, 8))
sns.violinplot(x='site', y='cong_pc', data=df)
plt.title('Congestion Percentage by Site')
plt.xlabel('Site')
plt.ylabel('Congestion Percentage')
plt.show()
```

```python
# Relationship between 'flow' and 'cong' using a joint plot
sns.jointplot(x='flow', y='cong', data=df, kind='scatter')
plt.suptitle('Joint Plot of Flow and Congestion', y=1.02)
plt.show()
```

```python
# Bar plot for average flow by day
plt.figure(figsize=(12, 6))
sns.barplot(x='day', y='flow', data=df, ci=None)
plt.title('Average Flow by Day')
plt.xlabel('Day')
plt.ylabel('Average Flow')
plt.show()
```

```python
# Boxplot for congestion percentage by day
plt.figure(figsize=(12, 6))
sns.boxplot(x='day', y='cong_pc', data=df)
plt.title('Congestion Percentage by Day')
plt.xlabel('Day')
plt.ylabel('Congestion Percentage')
plt.show()
```
